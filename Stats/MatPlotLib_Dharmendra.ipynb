```python
# Draw line graph
# include plot immediately after code
%matplotlib inline 
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-3, 3, 0.1)
y = norm.pdf(x) # find propbablities of x as per normal distribution

plt.plot(x,y)
# show x,y values
plt.show() #to display all figures
print(x.std())
print(np.mean(x))
```
