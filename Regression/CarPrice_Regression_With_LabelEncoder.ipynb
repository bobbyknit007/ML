```python
def remove_outliers(data_frame, series_col):
 Q1 = np.percentile(series_col, 25)
 Q3 = np.percentile(series_col, 75)
 IQR = Q3 - Q1
 Lower_Limit = Q1 - 1.5 * IQR
 Upper_Limit = Q3 + 1.5 * IQR
 data_frame = data_frame[(series_col >Lower_Limit) & (series_col < Upper_Limit)]
 return data_frame
```

```python
X =np.array(df['car_age']).reshape(-1,1)
Y =np.array(df['selling_price']).reshape(-1,1)
X_test,X_train,Y_test,Y_train =train_test_split(X,Y,test_size=0.2)
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
r2=r2_score(Y_test, y_pred)
print(r2)
```

When Linear Regression is don with feature variable car_age, the r2 score /accuracy is 40.8%
