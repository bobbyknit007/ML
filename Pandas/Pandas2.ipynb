```python
data = pd.DataFrame({'Brand': ['Maruti', 'Hyundai', 'Tata', 
                               'Mahindra', 'Maruti', 'Hyundai', 
                               'Renault', 'Tata', 'Maruti'], 
                     'Year': [2012, 2014, 2011, 2015, 2012, 
                              2016, 2014, 2018, 2019], 
                     'Kms Driven': [50000, 30000, 60000, 
                                    25000, 10000, 46000, 
                                    31000, 15000, 12000], 
                     'City': ['Gurgaon', 'Delhi', 'Mumbai', 
                              'Delhi', 'Mumbai', 'Delhi', 
                              'Mumbai', 'Chennai',  'Ghaziabad'], 
                     'Mileage':  [28, 27, 25, 26, 28, 
                                  29, 24, 21, 24]}) 
data
```

```python
# Display Cars with Brand Maruti and Mileage >25
#type((data.Brand == 'Maruti') & (data.Mileage >25)) # return 1-D array Series 
#data.loc[(data.Brand == 'Maruti') & (data.Mileage >25)] # loc method takes names/labels and bool , inlcudes last element

# update data , update Mileage column to value 22 when Year < 2025
data.loc[(data.Year < 2015), ['Mileage']] =22
data
```

```python
#iloc is index based, does not accept bool
data.iloc[[0,2,4,7]]
```

```python
data.iloc[1:5, 2: 5] # rows 1 to 4 and columns 2 to 4
```

```python
# at - position of element , column name
# iat - position of element in row, position of element in column
print(data.at[6, 'Kms Driven'])
print(data.iat[6,2])
```

```python
# query - takes expression in string , the expression must evlaute to bool
data.query('Brand  == "Tata" & Year > 2015')
```
