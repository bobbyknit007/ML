```python
#Drop The Column educational-num,capital-gain,capital-loss
df.drop(['educational-num','capital-gain','capital-loss'], axis=1, inplace=True)
df.head()
```

```python
#Find the most common education level
df['education'].value_counts().idxmax()# returns the index label of max value
```

```python
#find Average hours worked for ">50K" income
df[df['income']=='>50K']['hours-per-week'].mean()
```

```python
#Percentage of females in "Machine-op-inspct" occupation
#df[df['occupation']=='Machine-op-inspct']
female_occu=df[df['gender']=='Female']['occupation'].value_counts()
per=(female_occu['Machine-op-inspct']/female_occu.values.sum())*100
print('{} percent females work as machine-op-inspct'.format(per))
```

```python
#find out the avg age for each occupation
df.groupby('occupation').apply(lambda se : se['age'].mean())

```

```python
#find out Most common race among "Never-married"

df[df['marital-status']=='Never-married']['race'].value_counts().sort_values(ascending=False).index[0]
```

```python
#Percentage of people with a college degree or higher
edu_list=['Some-college','Bachelors','Masters','Doctorate']
edu_series =df.groupby('education')['education'].apply(lambda a :a).value_counts()
edu_degree = edu_series[edu_series.index.isin(edu_list)]
total=edu_series.values.sum()
print('Percentage of people with a college degree or higher {}'.format((edu_degree/total*100).sum()))

```
